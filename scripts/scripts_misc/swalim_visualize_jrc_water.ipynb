## Load packages and initialize GEE

```python
# Import Dependencies
import ee, datetime
import fiona
import pandas as pd
import matplotlib.dates as mdates
from IPython.display import Image
from matplotlib import dates
from pylab import *
ee.Initialize()


print('Packages loaded')
```

```python
water = ee.Image("JRC/GSW1_0/GlobalSurfaceWater");
aoi   = ee.FeatureCollection("users/remidannunzio/MGRS_100kmSQ_ID_38N");

water_aoi = water.clip(aoi);

Image(url=water_aoi.getThumbUrl({
    'region':aoi.geometry().bounds().getInfo()['coordinates']
}))
```

```python
task = ee.batch.Export.image.toAsset(image = water_aoi,
                                      description = 'water 38 NNJ',
                                      assetId = 'users/remidannunzio/jrc_water',
                                      scale = 30,
                                      region = aoi.geometry().bounds().getInfo()['coordinates'],
                                      maxPixels=1e12
                                     );

task.start()


state = task.status()['state']
while state in ['READY', 'RUNNING']:
    print(state + '...')
    time.sleep(5) 
    state = task.status()['state']
print('Done.', task.status())
print(task.status()['state'])
```
